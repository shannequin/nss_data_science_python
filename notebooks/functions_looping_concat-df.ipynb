For this warmup, you've been provided four csv files containing sales data by region for the years 2019 through 2022 (e.g., sales_2019.csv, sales_2020.csv, etc.).  
Each file includes columns for region and sales, but no year column.

```python
import pandas as pd
```

```python
pd.read_csv("../data/sales_2019.csv")
```

**Part 1:** Write a function that takes as input a year, which reads in the sales data for the input year, adds a Year column, and returns the resulting dataframe.

```python
def add_year_column(year: str) -> "DataFrame":
    # FILE PATH PATTERN
    file_path = f"../data/sales_{year}.csv"

    # READ FILE
    sales_df = pd.read_csv(file_path)

    # ADD YEAR COLUMN
    sales_df['year'] = year

    return sales_df
```

**Part 2:** Use your function in a loop or list comprehension to read in all of the sales data and combine it into a single dataframe.

```python
# INITIALIZE DATAFRAME
sale_year_df = pd.DataFrame()

# LOOP THROUGH REQUIRED YEARS
for year in range(2019, 2023):
    # CALL THE FUNCTION AND COMBINE THE DATAFRAMES
    sale_year_df = pd.concat([sale_year_df, add_year_column(year=str(year))])

# RESET THE INDEX
sale_year_df.reset_index(drop=True)
sale_year_df
```

**Part 3:** Which year had the highest total sales?

```python
sale_year_df.groupby("year")['sales'].sum().sort_values(ascending=False).head(1)
```

**Part 4:** Which region had the highest total combined sales across all 4 years?

```python
sale_year_df.groupby("region")['sales'].sum().sort_values(ascending=False).head(1)
```
