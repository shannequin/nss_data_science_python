## Part 1

You've been tasked with cleaning the results from a customer feedback survey. Customers were asked to provide their response as a number between 1 and 10. However, the form was free-text, so some people may have entered numbers, words, or even symbols. 

For the purposes of this exercise, we'll ignore any values that a non-numeric in calculating this average.

To start, here is a feedback list. Write a loop to process this list and output the average of the numeric entries.

**Hint:** You may want to make use of the [isinstance function](https://www.w3schools.com/python/ref_func_isinstance.asp).

```python
# Expected results: 7.16667

feedback = [8, "Good", 6.0, "N/A", 9, "Average", 10, 3, "Poor", 7]

def get_average_feedback(feedback: list) -> float:
    integer_list = []
    for value in feedback:
        if not isinstance(value, str):
            integer_list.append(value)
    
    return sum(integer_list) / len(integer_list)

get_average_feedback(feedback=feedback)
```

## Part 2

In the data folder is a file, store_feedback.json. It contains the feedback for a few different stores. Write some code to find the average score by store. Which store has the highest average score?

**Hint:** You may want to turn the code you used on Part 1 into a function that you can reuse for this part.

```python
import json

with open('../data/store_feedback.json', 'r') as file:
    feedback_list_stores = json.load(file)
```

```python
highest_score = 0
highest_store = ""

# lOOP THE FEEDBACK_LIST_STORES TO GET STORES
for store in feedback_list_stores:

    # GET AVERAGE FEEDBACK SCORES
    average_score = get_average_feedback(feedback=store['feedback'])
    store['average_score'] - average_score

    if average_score > highest_score:
        highest_score = average_score
        highest_store = store['store_number']
        
        
print(f"HIGHEST AVERAGE SCORE: {highest_score} for store {highest_store}")
```
