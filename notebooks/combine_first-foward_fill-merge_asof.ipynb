## Warmup: Time Since Last Maintenance 

**Scenario:** A factory tracks breakdowns of a machine and logs routine maintenance events.

Your goal is to find, for each breakdown, the time since the most recent routine maintenance.

```python
import pandas as pd

breakdowns = pd.DataFrame({
    "event_id": [1, 2, 3],
    "timestamp": pd.to_datetime([
        "2024-02-01 08:05",
        "2024-02-02 13:15",
        "2024-02-05 09:45",
    ])
})

maintenance = pd.DataFrame({
    "maintenance_id": [101, 102, 103, 104],
    "completed_at": pd.to_datetime([
        "2024-01-31 16:30",
        "2024-02-01 07:50",
        "2024-02-03 10:00",
        "2024-02-05 08:00",
    ])
})
```

## Step 1: The Difficult Way (using concat)

First, use [pd.concat](https://pandas.pydata.org/docs/reference/api/pandas.concat.html) to concatenate the two tables. 

Create a new column, 'sort_column' by taking either the value from the timestamp or the completed_at columns, whichever is non-null. Then sort by this column.

Hint: Use the [combine_first method](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.combine_first.html).

To get time of last maintenance event, [forward-fill](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.ffill.html) the completed_at column. 

Filter down to the breakdown rows.

Finally, calculate the time since maintenance by subtracting the completed_at column from the timestamp column.

## Step 2: The Better Way 

Use [merge_asof](https://pandas.pydata.org/docs/reference/api/pandas.merge_asof.html) to attach the most recent maintenance completion to each breakdown event. 

Then use the resulting DataFrame to calculate the time since maintenance.
