```python
import pandas as pd
```

```python
animals = pd.read_csv("../data/animals.csv")
animals.head()
```

```python
animals['Type'].value_counts()
```

```python
animals.groupby('Type')['Days in Shelter'].mean()
animals.groupby('Type').agg({'Days in Shelter':'mean'})
```

```python
animals.groupby('Type')['Outcome Type'].describe()
```

```python
animals[animals['Type']=='OTHER']['Breed'].unique()
```

```python
animals.groupby("Breed")['Days in Shelter'].mean().sort_values(ascending=False).head(1)
```

```python
animals_filtered = animals['Breed'].value_counts()
breed_filter = animals_filtered[animals_filtered >= 100].index.to_list()
animals[animals['Breed'].isin(breed_filter)].groupby('Breed')['Days in Shelter'].mean().sort_values()
```

```python
animals['Breed'].value_counts()[lambda x: x >= 100]
```

```python
animals.groupby('Breed')['Days in Shelter'].agg(['mean', 'count'])[lambda row: row['count'] >= 100].sort_values('mean')
```

```python
animals["Outcome Type"].value_counts(normalize=True).loc['ADOPTION']
```

```python
pd.crosstab(animals['Breed'], animals['Outcome Type'], normalize="index")['ADOPTION'].nlargest(1, keep="all")
```

```python
animals['adopted'] = animals['Outcome Type'] == "ADOPTION"
animals.groupby("Breed")['adopted'].mean()
```

```python
animals.groupby("Breed")['Outcome Type'].agg(lambda outcome: (outcome == "ADOPTION").mean())
```
