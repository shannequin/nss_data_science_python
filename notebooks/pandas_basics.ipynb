#### Modifying

You can change column names using the rename method.

```python
penguins.rename(columns={'year': 'observation_year'})
```

```python
penguins.head(1)
```

#### Slicing and Filtering

We can use .loc to **filter** our DataFrame to rows of interest.

A common way of doing this is to write a condition describing the rows that we want to keep. For example, if we want to look at rows for the chinstrap species.

```python
penguins['species'] == 'Chinstrap'
```

We can then pass this to .loc.

```python
penguins.loc[penguins['species'] == 'Chinstrap']
```

You can also specify which columns to keep. In general you will do

df.loc[rows to keep, columns to keep]

```python
penguins.loc[penguins['species'] == 'Chinstrap', ['body_mass_g', 'sex']]
```

Or to keep all rows and just those two columns.

```python
penguins.loc[:, ['body_mass_g', 'sex']]
```

It is also common to select columns using just double square brackets.

```python
penguins[['body_mass_g', 'sex']]
```
